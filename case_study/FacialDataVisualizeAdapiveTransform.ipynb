```python
import numpy as np
import torch
from torchvision import transforms
import pickle
from torchvision.utils import save_image
import warnings
warnings.filterwarnings('ignore')
```

# Visualize Transform

```python
X_tensor, y_tensor = pickle.load(open('../data/facial_attribute/fairface_set1_tensor.pkl', 'rb'))
```

```python
torch.manual_seed(2022)
trans1 = transforms.GaussianBlur(kernel_size=15)
trans2 = transforms.ColorJitter(brightness=0.2, contrast=0.2, saturation=0.2, hue=0.2)
trans_all = transforms.Compose([trans1, trans2])
```

```python
np.random.seed(2022)
select_idx = np.random.permutation(len(X_tensor))

img_list = []
for i in range(60):
    img_list.append(trans1(trans2(X_tensor[select_idx[i]])))
imgs = torch.stack(img_list)

save_image(imgs, "samples.png", nrow=10)

img_list = []
for i in range(60):
    img_list.append(X_tensor[select_idx[i]])
imgs = torch.stack(img_list)

save_image(imgs, "org_samples.png", nrow=10)
```
